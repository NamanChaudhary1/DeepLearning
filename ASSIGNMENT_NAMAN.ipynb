P1: a) Optimizing the Equation using Gradient Descent

```python
import numpy as np

# Define the function and its derivative
def f(x):
    return x**2 + 4*x + 4

def df(x):  # Derivative of f(x)
    return 2*x + 4

# Gradient Descent algorithm
def gradient_descent(learning_rate, n_iterations, initial_x):
    x = initial_x
    for i in range(n_iterations):
        gradient = df(x)
        x = x - learning_rate * gradient
        if i % 10 == 0:
            print(f"Iteration {i}: x = {x}, f(x) = {f(x)}")
    return x

# Parameters
learning_rate = 0.1
n_iterations = 100
initial_x = 10

# Finding the minima
minima = gradient_descent(learning_rate, n_iterations, initial_x)
print(f"Minima of the function is at x = {minima}")
```

P1: c) Random Data Arrays and Profit Calculation

```python
np.random.seed(0)  # Set seed for reproducibility
revenue = np.random.randint(10000, 50000, 15)
expenses = np.random.randint(9000, 27000, 15)

# Calculate Profit, Tax, and Margin
profit = revenue - expenses
tax = 0.3 * revenue
profit_after_tax = profit - tax
margin = (profit_after_tax / revenue) * 100

# Find good, bad, and worst months
mean_profit = np.mean(profit_after_tax)
good_months = np.where(profit_after_tax > mean_profit)[0]
bad_months = np.where(profit_after_tax < mean_profit)[0]
worst_month = np.argmin(profit_after_tax)

# Print results
print(f"Revenue: {revenue}")
print(f"Expenses: {expenses}")
print(f"Profit: {profit}")
print(f"Profit After Tax: {profit_after_tax}")
print(f"Margin: {margin}")
print(f"Good Months: {good_months}")
print(f"Bad Months: {bad_months}")
print(f"Worst Month: {worst_month}")
```

P2: a) Curve Fitting

```python
from scipy.optimize import curve_fit

# Example data: Processor Speed vs Price
processor_speed = np.array([1.4, 1.6, 1.8, 2.0, 2.2, 2.4, 2.6, 2.8, 3.0, 3.2])
price = np.array([100, 150, 200, 240, 290, 340, 390, 440, 500, 560])

# Define a model (linear in this case)
def model(x, a, b):
    return a * x + b

# Curve fitting
params, covariance = curve_fit(model, processor_speed, price)
print(f"Fitted parameters: {params}")

# Plot the fitted curve
plt.scatter(processor_speed, price, label='Data')
plt.plot(processor_speed, model(processor_speed, *params), label='Fitted Curve', color='red')
plt.xlabel('Processor Speed (GHz)')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

P2: b) Neural Network and Regression

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# Generate random data
x = np.random.rand(200)
y = 2 * x + 3 + np.random.normal(0, 0.2, 200)

# Plot data points
plt.scatter(x, y)
plt.title('Random Data Points')
plt.show()

# Neural network
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=1))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))

# Compile model
model.compile(optimizer='sgd', loss='mse')

# Train the model
history = model.fit(x, y, epochs=100, verbose=0)

# Plot loss function vs epochs
plt.plot(history.history['loss'])
plt.title('Loss Function vs No of Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

P2: c) Compare Optimizers

```python
# Using SGD, Mini-batch SGD, and Adam
from tensorflow.keras.optimizers import SGD, Adam

optimizers = ['sgd', 'adam', 'RMSprop']

# Plot comparison for each optimizer
for opt in optimizers:
    model.compile(optimizer=opt, loss='mse')
    history = model.fit(x, y, epochs=50, verbose=0)
    plt.plot(history.history['loss'], label=opt)

plt.title('Loss Function Descent for Different Optimizers')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

ANN Regression for Curve Fit

```python
import numpy as np
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
# Generate synthetic data
X = np.linspace(-5, 5, 100).reshape(-1, 1)
y = X**3 - 2 * X**2 + 3

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# ANN Model
model = MLPRegressor(hidden_layer_sizes=(50,), max_iter=1000)
model.fit(X_train, y_train.ravel())

# Predictions
y_pred = model.predict(X_test)

# Plot
plt.scatter(X_test, y_test, label="Actual Data")
plt.scatter(X_test, y_pred, label="Predicted Data", color='red')
plt.legend()
plt.show()
```

Using ANN as a Classifier on Clusters of Random Points

```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt

# Generate random clusters
X, y = make_blobs(n_samples=300, centers=3, random_state=42)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Generated Clusters")
plt.show()

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# ANN Classifier
clf = MLPClassifier(hidden_layer_sizes=(50,), max_iter=1000)
clf.fit(X_train, y_train)

# Accuracy
accuracy = clf.score(X_test, y_test)
print(f"Accuracy: {accuracy:.2f}")
```

 Binary Classification of Random Circular Data

```python
from sklearn.datasets import make_circles
from sklearn.svm import SVC

# Generate circular data
X, y = make_circles(n_samples=300, noise=0.05, factor=0.5)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
plt.title("Circular Data")
plt.show()

# Train SVM classifier
svc = SVC(kernel='rbf', gamma=1)
svc.fit(X, y)

# Plot decision boundary
import numpy as np
xx, yy = np.meshgrid(np.linspace(-1.5, 1.5, 500), np.linspace(-1.5, 1.5, 500))
Z = svc.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.contourf(xx, yy, Z, levels=10, cmap='coolwarm', alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
plt.title("Decision Boundary")
plt.show()
```

Implementation of LeNet CNN for MNIST Dataset

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, AveragePooling2D, Flatten, Dense, InputLayer

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# LeNet Model
model = Sequential([
    InputLayer(input_shape=(28, 28, 1)),
    Conv2D(6, kernel_size=(5, 5), activation='tanh', padding='same'),
    AveragePooling2D(pool_size=(2, 2)),
    Conv2D(16, kernel_size=(5, 5), activation='tanh'),
    AveragePooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(120, activation='tanh'),
    Dense(84, activation='tanh'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Train the model
model.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_test, y_test))
```

Generate MNIST Digits Using Deep Autoencoders or GANs

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt

# Define autoencoder
input_img = Input(shape=(784,))
encoded = Dense(128, activation='relu')(input_img)
encoded = Dense(64, activation='relu')(encoded)
encoded = Dense(32, activation='relu')(encoded)
decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(128, activation='relu')(decoded)
decoded = Dense(784, activation='sigmoid')(decoded)

autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# Load and preprocess MNIST
(x_train, _), (x_test, _) = tf.keras.datasets.mnist.load_data()
x_train = x_train.reshape(-1, 784) / 255.0
x_test = x_test.reshape(-1, 784) / 255.0

# Train autoencoder
autoencoder.fit(x_train, x_train, epochs=10, batch_size=256, shuffle=True, validation_data=(x_test, x_test))

# Visualize reconstructed digits
reconstructed = autoencoder.predict(x_test[:10])
plt.figure(figsize=(10, 4))
for i in range(10):
    plt.subplot(2, 10, i + 1)
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.subplot(2, 10, i + 11)
    plt.imshow(reconstructed[i].reshape(28, 28), cmap='gray')
plt.show()
```

CNNs for CIFAR-10 Classification

```python
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
y_train, y_test = to_categorical(y_train), to_categorical(y_test)

# CNN Model
cnn_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
cnn_model.summary()

# Train the model
cnn_model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_test, y_test))
```
